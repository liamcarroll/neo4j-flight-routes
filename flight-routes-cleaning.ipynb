```python
#Imports

import pandas as pd
```

```python
#Reading in airports in EU and US, which are in the top 5 airports in the country. 

df_airports = pd.read_csv(r'airports_final.csv', sep=',', header=None, names=['Airport_ID', 'Name', 'City', 'Country', 'IATA_code', 'ICAO_code', 'Lat', 'Long', 'Altitude', 'Timezone', 'DST', 'Timezone_formatted', 'type', 'source'], index_col=0, skiprows=1)
```

```python
df_airports.head()
```

```python
df_airports.columns
```

```python
airport_codes = set(df_airports['IATA_code'])
```

```python
'BRU' in airport_codes
```

```python
#Importing routes dataset.

df_routes = pd.read_table(r'routes.dat', sep=',', header=None, names=['Airline', 'Airline_ID', 'Source_Airport', 'Source_Airport_ID', 'Dest_Airport', 'Dest_Airport_ID', 'Codeshare', 'Stops', 'Equipment'])
```

```python
df_routes.head()
```

```python
df_routes.shape
```

```python
#Only routes whose source airport is in our list

df_routes_filtered = df_routes[df_routes['Source_Airport'].isin(airport_codes)]
```

```python
df_routes_filtered.shape
```

```python
#Narrow this again to only routes whose source and dest are in our list

df_routes_filtered = df_routes_filtered[df_routes_filtered['Dest_Airport'].isin(airport_codes)]
```
